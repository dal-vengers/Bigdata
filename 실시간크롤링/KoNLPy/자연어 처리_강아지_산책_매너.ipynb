```python
import konlpy
from konlpy.tag import Kkma

kkma = Kkma()

textNouns =list()

for i in range(0,2088):
    textNoun = kkma.nouns(texts[i])
    textNouns.append(textNoun)

print(textNouns)
```

```python
finalTexts=list()
for i in range(0,len(textNouns)):
    for j in range(0,len(textNouns[i])):
        finalTexts.append(textNouns[i][j])
print(finalTexts)
```

#### 불용어 출처 : https://www.ranks.nl/stopwords/korean

```python
stop_words = list()

f1 = open("/Users/juheekim/Desktop/coding/mycodesource/Multicampus/4조/mine/자연어처리/한국어불용어_강아지_산책_매너.csv",'r')
rea1 = csv.reader(f1)
for row in rea1:
    stop_words.append((row[0]))

print(stop_words)
```

```python
keywords=['강아지', '산책', '매너']
```

```python
textDict={}

for word in finalTexts:
    if word not in stop_words:
        if len(word)>1 or word in ["똥","풉","변","떵"]:
            if word not in keywords:
                if word not in textDict:
                    textDict[word]=1
                else:
                    textDict[word]+=1
print(textDict)
```

```python
textDict_sorted = sorted(textDict.items(), key=lambda x:x[1], reverse=True)
print(textDict_sorted)
```

```python
word_index={}
i=0
for (word, freq) in textDict_sorted:
    if freq >20 :
        i+=1
        word_index[word] = i

print(word_index)
```

```python
top20=textDict_sorted[0:21]
print(top20)
```

```python
import pandas as pd

df_top20=pd.DataFrame(columns=['keywords','freq'])

for i in range(0,20):
    df_top20.loc[i]=[top20[i][0],top20[i][1]]

display(df_top20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='darkgrid')

fig = plt.figure(figsize=(50, 12))   
ax1 = fig.add_subplot(1, 3, 1)
ax1.set_title('강아지 산책 매너')
plt.xticks(rotation=45)
sns.barplot(x='keywords', y='freq', data=df_top20)
plt.show()
```
